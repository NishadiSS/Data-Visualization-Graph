```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random 2D data (10x10 array)
data = np.random.rand(10, 10)  # Generate a 10x10 array of random numbers between 0 and 1

# Plotting the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(data, cmap='viridis', interpolation='nearest')  # Plot the heatmap with 'viridis' colormap
plt.colorbar()  # Add color bar to show scale
plt.title('Heatmap of Random 2D Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
