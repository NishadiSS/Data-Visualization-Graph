```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd  # Optional for handling time series data

# Generate synthetic time series data
np.random.seed(0)  # Set random seed for reproducibility
dates = pd.date_range('2022-01-01', periods=365)  # Create a range of dates for one year (daily frequency)
values = np.cumsum(np.random.randn(365))  # Generate random data (cumulative sum of random values)

# Plotting the time series
plt.figure(figsize=(10, 6))
plt.plot(dates, values, linestyle='-', color='blue', linewidth=2)  # Plot the time series
plt.title('Time Series Plot')
plt.xlabel('Date')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
