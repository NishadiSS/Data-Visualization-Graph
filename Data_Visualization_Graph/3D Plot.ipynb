```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate random 3D data (x, y, z coordinates)
np.random.seed(0)
num_points = 100
x = np.random.rand(num_points)  # Random x values
y = np.random.rand(num_points)  # Random y values
z = np.random.rand(num_points)  # Random z values

# Plotting the 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with random colors
ax.scatter(x, y, z, c=z, cmap='viridis', s=100, alpha=0.8, edgecolors='w')

# Customize labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('3D Scatter Plot')
plt.show()
```
