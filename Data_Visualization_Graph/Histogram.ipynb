```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data (normal distribution)
data = np.random.randn(1000)  # Generate 1000 random data points from a standard normal distribution

# Plotting the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', alpha=0.7)  # Plot histogram with 30 bins, skyblue color, and transparency
plt.title('Histogram of Random Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```
